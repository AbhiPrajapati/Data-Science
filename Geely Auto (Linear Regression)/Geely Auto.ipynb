# Univariate Analysis

A general function for univariate analysis for all categorical variables in the data set.

```python
from tabulate import tabulate

def univariate_categorical(df):
        
   cat_cols = [cname for cname in df.columns if df[cname].dtype == "object"]   

   for col in cat_cols:
    
       len_cat = len(df[col].unique())
    
       print('\n############ ' + col + ' ############')
       print('\nNumber of unique values => ' + str(len_cat) + '\n\n')

       if len_cat > 10:
           plt.figure(figsize=(15, 12))
            
       else:
           plt.figure(figsize=(10, 8))
        
       y = "count"
       plt.subplot(1, 2, 1)
       sns.countplot(df[col])
       count_df = df[col].value_counts().rename(y).reset_index().rename(columns={"index":col})

       y = "percent(%)"
       percent_df = df[col].value_counts(normalize=True).apply(lambda x: round(x*100, 2)).rename(y).reset_index().rename(columns={"index":col})
       plt.subplot(1, 2, 2)
       sns.barplot(percent_df[col], percent_df[y], data=percent_df)

       plt.show()
       print(tabulate(pd.merge(percent_df, count_df, how='inner'), headers='keys', tablefmt='fancy_grid'))
```

```python
univariate_categorical(df)
```

A general function for segmented univariate analysis for all categorical variables in the data set against price, the target variable.

```python
def segmented_univariate_all(df, target):

    cat_cols = [cname for cname in df.columns if df[cname].dtype == "object"]

    for col in cat_cols:
    
        print('\n####################### ' + col + ' #######################')
        print('\nNumber of unique values => ' + str(len(df[col].unique())) + '\n\n')

        # Create 2 subplots for mean and median respectively
        plt.figure(figsize=(12, 8))

        # subplot 1: statistic=mean
        plt.subplot(1, 2, 1)
        sns.barplot(x=col, y=target, data=df)
        plt.title("Mean")

        # subplot 2: statistic=median
        plt.subplot(1, 2, 2)
        sns.barplot(x=col, y=target, data=df, estimator=np.median)
        plt.title("Median")

        plt.show()
```

Defining general function to calculate the VIF of the variables in the model.

```python
def VIF(X):
    # Calculate the VIFs for the new model
    from statsmodels.stats.outliers_influence import variance_inflation_factor

    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    print(vif)
```

Defining a general linear model fit function.

```python
def lm_fit(X):

    # Adding a constant variable 
    import statsmodels.api as sm  
    X_train_lm = sm.add_constant(X)

    lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model

    #Let's see the summary of our linear model
    print(lm.summary())
```

2) The price is increased by 0.23 times when all the other variables are constant and the car company is either BMW or Porsche.

3) The price is increased by 0.03 times when all the other variables are constant and the aspiration of the car is turbo.

4) The price is increased by 0.26 times when all the other variables are constant and the enginetype of the car is rotor.
